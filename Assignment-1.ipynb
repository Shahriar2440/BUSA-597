```python
#Removing the ouliners
# Max = 75% + 1.5*(75%-25%)
max_wind = 26.825 + 1.5*(26.825-17.4)

new_wind_speed = weather[weather['windspeed'] <= max_wind]
new_wind_speed.plot.box()
```

```python
# Q-2: Correlation

Dal_weather_cor = new_wind_speed.corr()
Dal_weather_cor
```

```python
# Visualization - Correlation
%matplotlib inline
import seaborn as sb
sb.heatmap(Dal_weather_cor)
```

```python
# Using the heatmap package, there were sevearl variables who were correlated with each other significantly especially more than
# 80%. For instance, 'tempmax', 'temp', 'fellslikemin', 'dew', 'solarenergy' and 'uvindex'.
# Remove Correlated variables

final_weather = new_wind_speed.drop(columns =['tempmax','feelslikemin','dew','precipcover'])
final_weather
```
